## Interactive FISS Data Viewer (IFDV) Usage

The Interactive FISS Data Viewer (IFDV) is the widget code to easily show the FISS data and interactively change the some parameters written in python using Qt5.

Here we breifly introduce this code.

# Usage in IPython pannel

There are two ways to use this code.

1. Input the directory which has the FISS data (only for comp or proc data)

```python
%matplotlib
```

```python
from fisspy import IFDV
```

```python
fdir=r'D:\Research\Data\FISS\FTS\140603\proc'
```

```python
IFDV(fdir=fdir)
```

In this notebook there is an error seen above line, but it is just the reason that notebook can not be open the new window.

So in your computer, there may be no error. If not, please change your matplotlib backend to qt5.

----------------------------

Then new matplotlib window is open, and there are 4 FISS data pannels and a control pannel at the bottom of window.

On the FISS data pannels, left two pannels show the raster images of FISS camera A and B, and right two pannel show the line profiles of each FISS cameras for the given position.

 * You can change the postion for line profile by pressing the 'q' button on the raster pannels, and also you can change the wavelength of the shown raster images by pressing 'q' button on the line profile pannels.
 * You can change the color range of two raster images by giving the value or by sliding the each sliders on the control pannel.
 * you can show other time data by inputting the frame number or sliding the time slider on the control pannel.

```python
from glob import glob
from fisspy import IFDV
```
